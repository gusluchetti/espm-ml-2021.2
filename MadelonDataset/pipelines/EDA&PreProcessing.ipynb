## Importing Training Dataset

```python
import pandas as pd

# importing training dataset
X_train = pd.read_csv("../dataset files/training/madelon_train.data", 
                      header=None, index_col=None, delimiter=' ')
X_train = X_train.drop(columns=[500])

X_train.info()
X_train.head()
```

```python
# describe for training dataset
X_train.describe()
```

## Normalizing dataset

Since this is a sintetic dataset, there's no harm in normalizing the data for every single pipeline. (no real loss of information over non-normalized data.)

```python
# normalizing training dataframe
from sklearn import preprocessing

min_max_scaler = preprocessing.MinMaxScaler()
X_train_scaled = pd.DataFrame(
    min_max_scaler.fit_transform(X_train), 
        index=X_train.index, 
        columns=X_train.columns
    )

X_train_scaled
```

## Exporting dataset values

```python
X_train_scaled.to_csv(path_or_buf="../dataset files/training/post_eda_train.csv")
```
